```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 
# plt.rcParams['font.sans-serif'] = ['SimHei']  # win
plt.rcParams['font.family'] = 'STHeiti' # mac

plt.rcParams['axes.unicode_minus'] = False    # 

# 
missiles = {
    'M1': (20000, 0, 2000),
    'M2': (19000, 600, 2100),
    'M3': (18000, -600, 1900)
}

# 
drones = {
    'FY1': (17800, 0, 1800),
    'FY2': (12000, 1400, 1400),
    'FY3': (6000, -3000, 700),
    'FY4': (11000, 2000, 1800),
    'FY5': (13000, -2000, 1300)
}

# 
targets = {
    '': (0, 0, 0),
    '': (0, 200, 0)
}

# 3D
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 
for name, pos in missiles.items():
    ax.scatter(pos[0], pos[1], pos[2], c='red', marker='^', s=100, label=f'{name}' if name == 'M1' else "")
    ax.text(pos[0], pos[1], pos[2], f'{name}', color='red')

# 
for name, pos in drones.items():
    ax.scatter(pos[0], pos[1], pos[2], c='blue', marker='o', s=100, label=f'{name}' if name == 'FY1' else "")
    ax.text(pos[0], pos[1], pos[2], f'{name}', color='blue')

# 
for name, pos in targets.items():
    ax.scatter(pos[0], pos[1], pos[2], c='green', marker='*', s=200, label=name)
    ax.text(pos[0], pos[1], pos[2], name, color='green')

# 
ax.set_xlabel('X (m)')
ax.set_ylabel('Y (m)')
ax.set_zlabel('Z (m)')

# 
ax.set_title('3D')

# 
ax.legend()

# 
ax.view_init(elev=20, azim=45)

# 
plt.tight_layout()
plt.show()
```
