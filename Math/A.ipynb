```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 设置中文字体
# plt.rcParams['font.sans-serif'] = ['SimHei']  # win
plt.rcParams['font.family'] = 'STHeiti' # mac

plt.rcParams['axes.unicode_minus'] = False    # 解决负号显示问题

# 导弹位置
missiles = {
    'M1': (20000, 0, 2000),
    'M2': (19000, 600, 2100),
    'M3': (18000, -600, 1900)
}

# 无人机位置
drones = {
    'FY1': (17800, 0, 1800),
    'FY2': (12000, 1400, 1400),
    'FY3': (6000, -3000, 700),
    'FY4': (11000, 2000, 1800),
    'FY5': (13000, -2000, 1300)
}

# 目标位置
targets = {
    '假目标': (0, 0, 0),
    '真目标': (0, 200, 0)
}

# 创建3D图形
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# 绘制导弹位置
for name, pos in missiles.items():
    ax.scatter(pos[0], pos[1], pos[2], c='red', marker='^', s=100, label=f'导弹{name}' if name == 'M1' else "")
    ax.text(pos[0], pos[1], pos[2], f'{name}', color='red')

# 绘制无人机位置
for name, pos in drones.items():
    ax.scatter(pos[0], pos[1], pos[2], c='blue', marker='o', s=100, label=f'无人机{name}' if name == 'FY1' else "")
    ax.text(pos[0], pos[1], pos[2], f'{name}', color='blue')

# 绘制目标位置
for name, pos in targets.items():
    ax.scatter(pos[0], pos[1], pos[2], c='green', marker='*', s=200, label=name)
    ax.text(pos[0], pos[1], pos[2], name, color='green')

# 设置坐标轴标签
ax.set_xlabel('X坐标 (m)')
ax.set_ylabel('Y坐标 (m)')
ax.set_zlabel('Z坐标 (m)')

# 设置标题
ax.set_title('导弹和无人机位置分布3D图')

# 添加图例
ax.legend()

# 调整视角以便更好地查看所有点
ax.view_init(elev=20, azim=45)

# 显示图形
plt.tight_layout()
plt.show()
```
