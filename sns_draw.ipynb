# scatterplot

```python
tips = pd.read_csv('tips.csv')

tips.head()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="time",style="time")
```

```python
# 从#108dc7到#ef8e38的渐变色，共6种
gradient_colors = [
    "#108dc7",   # 起始色：亮蓝色
    "#4aa5c5",   # 渐变1/5处
    "#82bcc3",   # 渐变2/5处
    "#b9d2c1",   # 渐变3/5处
    "#f1e8bf",   # 渐变4/5处
    "#ef8e38"    # 结束色：橙色
]
sns.scatterplot(data=tips, x="total_bill", y="tip",hue="size",palette=gradient_colors)
```

```python
tip_rate = tips.eval("tip / total_bill").rename("tip_rate")# 计算小费比例
sns.scatterplot(data=tips, x="total_bill", y="tip", hue=tip_rate)
sns.regplot(data=tips, x="total_bill", y="tip", scatter=False, color="red")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size="size")# 设置点的大小
```

```python
sns.scatterplot(
    data=tips, x="total_bill", y="tip", hue="size", size="size",
    sizes=(20, 200), legend="auto"
)
# sizes: 控制散点图中点的大小范围，元组形式 (min_size, max_size)
# legend: 控制图例的显示方式，"full"表示显示所有图例项
# legend: 可选参数包括 "auto"（自动选择显示方式）、"brief"（简化显示）、False（不显示图例）
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
sns.scatterplot(data=tips, x="total_bill", y="tip", style="time", markers=markers)
```

# displot 分布图

```python
# The default plot kind is a histogram:
penguins = pd.read_csv("penguins.csv")

sns.displot(data=penguins, x="flipper_length_mm")
```

使用 kind 参数选择不同的表示形式：   
‘ecdf’ : 累积分布函数.

'hist' : 直方图. 
 
'kde' : 密度图. 

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", kind="ecdf")
```

```python
# 在直方图模式下，也可以添加 KDE 曲线：
sns.displot(data=penguins, x="flipper_length_mm", kde=True)
```

## 绘制二维的 双变量分布图

```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm", kind="hist")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", y="bill_length_mm",kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(data=penguins, x="flipper_length_mm", hue="species", col="sex", kind="kde")
```

# catplot 分类图


参数值	图表类型	描述
"strip"	带状图	显示所有数据点，适合小规模数据。
"swarm"	蜂群图	数据点无重叠排列，适合展示分布和密度。
"box"	箱线图	显示数据的五数概括（最小值、Q1、中位数、Q3、最大值）和异常值。
"violin"	小提琴图	结合箱线图和核密度估计，展示数据的分布形状。
"bar"	条形图	显示分类变量的均值或其他统计量。
"point"	点图	用点和线段表示统计量，适合比较不同组别的差异。
"count"	计数图	显示每个分类的观测数量。

```python
df = pd.read_csv('data/titanic.csv')
df.head()
```

```python
sns.catplot(
    data=df, x="age", y="class", hue="sex",
    kind="violin", bw_adjust=0.5, cut=0, split=True,
)
# bw_adjust: 控制核密度估计的带宽调整因子，值越大表示更平滑的曲线
# cut: 控制小提琴图的边界裁剪范围，0表示裁剪超出的部分
# split: 是否将小提琴图按hue分组拆分显示，True表示拆分
```

```python
sns.catplot(
    data=df, x="class", y="survived", col="sex",
    kind="bar", height=4, aspect=.6,
    # ci=None #控制误差线条
)
```

# rugplot 加上边缘密度地毯

```python
import seaborn as sns
import pandas as pd
sns.set_theme()
tips = pd.read_csv("./data/tips.csv")
```

```python
fig, axes = plt.subplots(2,3, figsize=(16, 10))
axes = axes.flatten()  # 将二维数组转为一维，便于循环访问
ax1 = axes[0]
ax2 = axes[1]
ax3 = axes[2]
ax4 = axes[3]
ax5 = axes[4]
ax6 = axes[5]
sns.kdeplot(data=tips, x="total_bill" ,ax=ax1)
sns.rugplot(data=tips, x="total_bill", ax=ax1)
ax1.set_title("Distribution of Total Bill")

sns.stripplot(data=tips, x="total_bill", y="day",ax=ax2)

sns.stripplot(data=tips, x="total_bill", y="day", hue="day", ax=ax3)

sns.stripplot(data=tips, x="total_bill", y="day", hue="size", ax=ax4)

sns.stripplot(
    data=tips, x="total_bill", y="day", hue="time",
    jitter=False, s=20, marker="D", linewidth=1, alpha=.1,
    ax=ax5
)
```

```python
sns.catplot(data=tips, x="time", y="total_bill", hue="sex", col="day", aspect=.5)
#aspect=.5 是一个控制图形宽高比的参数
```
