```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('_mpl-gallery')

# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
plt.figure(figsize=(10, 5))
fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# 创建一个 3x2 的子图布局
fig, axes = plt.subplots(2,3, figsize=(16, 8))

# 遍历每个子图并绘制内容
for i in range(2):
    for j in range(3):
        ax = axes[i, j]# 获取当前子图的坐标轴对象，axes[i, j]表示第i行第j列的子图

        ax.plot([0, 1], [0, 1])  # 示例：绘制一条直线
        ax.set_title(f'Subplot {i+1}-{j+1}')

# 调整子图间距
# plt.tight_layout()
plt.show()
```
