```python
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.family'] = 'STHeiti'
```

```python
#创建图标
plt.figure(figsize=(10,5))

month = ['1月','2月','3月','4月']
sales = [100,80,160,200]

plt.plot(month,sales,label="产品A",color='black')

#添加标签
plt.title("2025年第一季度销售趋势",fontsize=20,color='green')
plt.xlabel("月份")
plt.ylabel("销售额（万元）")
#添加图例
plt.legend(loc="upper left")
#添加网格线
plt.grid(axis='y',alpha=0.5)
#设置刻度字体大小
plt.xticks(rotation=0,fontsize=10)
#设置y轴的范围
plt.ylim(0,210)
#在每个点上显示数据
for x,y in zip(month,sales):
    plt.text(x,y+10,str(y),ha='center',va='bottom')
plt.show()
```
