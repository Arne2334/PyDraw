```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
np.random.seed(42)
n = 100
x = np.random.rand(n)
y = np.random.rand(n)
z = np.random.rand(n)

# 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 
scatter = ax.scatter(x, y, z, c=z, cmap='viridis', marker='o')

# 
plt.colorbar(scatter)

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 
t = np.linspace(0, 20, 1000)
x = np.sin(t)
y = np.cos(t)
z = t

# 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# 
ax.plot(x, y, z, 'b-', linewidth=2)

# 
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D')

plt.show()
```

```python
# mayavi: pip3 install mayavi

from mayavi import mlab
import numpy as np

# 
x, y = np.mgrid[-5:5:100j, -5:5:100j]
z = np.sin(np.sqrt(x**2 + y**2))

# 3D
mlab.figure(size=(800, 600))
s = mlab.surf(x, y, z, colormap='viridis')

# 
mlab.colorbar(s, title='', orientation='vertical')
mlab.xlabel('X')
mlab.ylabel('Y')
mlab.zlabel('Z')
mlab.title('Mayavi 3D')

mlab.show()
```
